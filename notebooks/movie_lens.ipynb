##### computing mean ratings for a particular movie by sex and age

#### Goal
+ create one unified table that includes, 
  + ratings - user ratings of the movies
  + users - information aout the user
  + movies - information about the movie
+ each row will have one rating, who gave it, which movie it was for, and details about the movie and the name
+ we want to get a full picture of every rating, who rated what movie, what they thought, and who they are

````python
# using pandas merge fn to merge  ratings with users then merging with the movies data
data = pd.merge(pd.merge(ratings, users), movies)
data.head()
````

```
data = pd.merge(pd.merge(ratings, users), movies)
```

`pd.merge(ratings, users)` - this merges the ratings dataframe with users on a common columns, ***user_id***

`pd.merge(..., movies)` - this merges the result of the first merge with the movies dataframe on the ***movie_id*** column which is common to both dataframes
