##### computing mean ratings for a particular movie by sex and age

#### Goal
+ create one unified table that includes, 
  + ratings - user ratings of the movies
  + users - information aout the user
  + movies - information about the movie
+ each row will have one rating, who gave it, which movie it was for, and details about the movie and the name
+ we want to get a full picture of every rating, who rated what movie, what they thought, and who they are

````python
# using pandas merge fn to merge  ratings with users then merging with the movies data
data = pd.merge(pd.merge(ratings, users), movies)
data.head()
````

```
data = pd.merge(pd.merge(ratings, users), movies)
```

`pd.merge(ratings, users)` - this merges the ratings dataframe with users on a common columns, ***user_id***

`pd.merge(..., movies)` - this merges the result of the first merge with the movies dataframe on the ***movie_id*** column which is common to both dataframes

````python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings[:5]
````

````python
mean_rating_s = pd.pivot_table(data, values='rating', index='title', columns='gender', aggfunc='mean')
mean_rating_s.head(10)
````

#### filter down to movies that received at least 250 ratings
+ this is done by using the pandas `groupby()` function
+ we group the data by title and use `size()` function to get a ***Series*** of group sizes for each title
+ `SeriesGroupBy.size()` - Compute group sizes

````python
ratings_by_title = data.groupby('title').size()
````

````python
ratings_by_title.head(10)
````

````python
ratings_by_title.info()
````

#### `rating_by_title` - this is a Pandas Series where the index is the movie title and the value is a count of ratings per title

````python
active_titles=ratings_by_title.index[ratings_by_title >= 250]
active_titles
````

##### `ratings_by_title >= 250` - this creates a boolean Series to check the popularity of the movies, we basically want to answer a question of whether the movie title got at least 250 ratings
##### `ratings_by_title.index[...]` - this does boolean indexing on the Series index itself and returns an Index object which is a list of movies that passed the `>= 250` filter
##### `active_titles` - this is a list like object that has popular movie titles that have ratings above 250
##### the object can now be used to slice down the full ratings table to those with enough data to be statistically interesting

````python
mean_ratings=mean_ratings.loc[active_titles]
mean_ratings
````

+ we filter the `mean_ratings` DataFrame to show the titles that are popular
+ now we have a clean, filtered table of movies that matter in the analysis

````python
top_female_ratings=mean_ratings.sort_values(by='F', ascending=False)
top_female_ratings[:10]
````

+ we sort the data frame by the female average rating highest first
+ the results of the line of code returns a data frame where women gave the highest ratings

##### measuring rating disagreement - the gender wars edition
+ the goal is to find the movies that are most divisive between male and female viewers

````python
mean_ratings['diff']=mean_ratings['M']-mean_ratings['F']
mean_ratings[:5]
````

+ we want to sort by the difference so that we know the movies with the greatest rating difference
+ that will tell us which movies were preferred by which gender
+ positive difference means men rated it higher
+ negative difference means women rated it higher

````python
sorted_by_diff=mean_ratings.sort_values(by='diff')
sorted_by_diff[:15]
````

````python
sorted_by_diff.tail(15)
````
